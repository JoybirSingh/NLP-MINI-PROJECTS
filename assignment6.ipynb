```python
import numpy as np
import pandas as pd

train= pd.read_csv("/content/AmesHousing.csv")
train.shape
train.head()
```

```python
train.drop(['PID'], axis=1, inplace=True)

df_test=pd.read_csv("/content/AmesHousing.csv")

df_test.shape
df_test.isna().sum()
```

```python
train['SalePrice'].describe()
```

```python
from tqdm import tqdm
test = tqdm(range(0, len(df_test)), desc='Matching')
for i in test:
    for j in range(0, len(train)):
        for k in range(1, len(df_test.columns)):
            if df_test.iloc[i,k] == train.iloc[j,k]:
                continue
            else:
                break
test.close()
```
